```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('enrichments_torus_ftest_nominal.txt', delimiter=r"\s+", skiprows=2, header = None,
                names = ['Annotation', 'Est', '2.5% conf', '97.5% conf'])
df.insert(4, 'Test', ['F-test' for i in range(len(df['Annotation']))])
df2 = pd.read_csv('enrichments_torus_qtltools_gene_nominal.txt', delimiter=r"\s+", skiprows=2, header = None,
                names = ['Annotation', 'Est', '2.5% conf', '97.5% conf'])
df2.insert(4, 'Test', ['QTLtools-sum' for i in range(len(df['Annotation']))])
df = df.append(df2)
df['Annotation'] = [i.split('.')[0] for i in df['Annotation']]
```

```python
df2 = df.drop(['Est', '97.5% conf'], axis=1)
df2 = df2.rename({'2.5% conf': 'Est'}, axis=1)
df3 = df.drop(['Est', '2.5% conf'], axis=1)
df3 = df3.rename({'97.5% conf': 'Est'}, axis=1)
df4 = df.drop(['2.5% conf', '97.5% conf'], axis=1)
df4 = df4.append(df2)
df4 = df4.append(df3)
df4.head()
```

```python
p = sns.pointplot(y='Annotation', x='Est', hue='Test', data=df4, dodge=0.3, join=False, scale=0.7, errwidth=1.5)
p.set(xlabel='log odds ratio', ylabel=None)
for y in range(0, len(df['Annotation'].unique())):
    p.axhspan(y - 0.5, y + 0.5, facecolor='black', alpha=[0.0 if y%2 == 0 else 0.1][0])
plt.tight_layout()
plt.savefig('enrichment_plot.png')
```
